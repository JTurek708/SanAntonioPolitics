```python
import pandas as pd
import calendar
import plotly.express as px
```

```python
df = pd.read_csv('campaign_finance20240926.csv')
df.head(), df.columns
```

```python
df['Election Date:'] = pd.to_datetime(df['Election Date:'], format='%m/%d/%y',
                                      errors='coerce')
df.head()
```

```python
df['TransDate:'] = df['TransDate:'].dt.normalize()
df['CreatedDt:'] = df['CreatedDt:'].dt.normalize()
df.head()
```

```python
# Create new cols for Election, Trans, Created calendar
df['Election Year'] = df['Election Date:'].dt.year
df['Election Month'] = df['Election Date:'].dt.month
df['Election Day'] = df['Election Date:'].dt.day

# Break down CreatedDt into Year, Month, Day
df['Created Year'] = df['CreatedDt:'].dt.year
df['Created Month'] = df['CreatedDt:'].dt.month
df['Created Day'] = df['CreatedDt:'].dt.day

# Break down TransDate into Year, Month, Day
df['Trans Year'] = df['TransDate:'].dt.year
df['Trans Month'] = df['TransDate:'].dt.month
df['Trans Day'] = df['TransDate:'].dt.day
```

```python
# Map month numbers to month names
df['Election Month'] = df['Election Month'].apply(lambda x: calendar.month_name[int(x)] if pd.notnull(x) else None)
df['Created Month'] = df['Created Month'].apply(lambda x: calendar.month_name[int(x)] if pd.notnull(x) else None)
df['Trans Month'] = df['Trans Month'].apply(lambda x: calendar.month_name[int(x)] if pd.notnull(x) else None)

# Convert year and day cols to integers
df['Election Year'] = df['Election Year'].astype('Int64')
df['Election Day'] = df['Election Day'].astype('Int64')
df['Created Year'] = df['Created Year'].astype('Int64')
df['Created Day'] = df['Created Day'].astype('Int64')
df['Trans Year'] = df['Trans Year'].astype('Int64')
df['Trans Day'] = df['Trans Day'].astype('Int64')

df.head()
```

```python
# Fix Manny Pelaez name
df['Cand/Committee:'] = df['Cand/Committee:'].replace({
    'Manny Pelaez Pelaez': 'Manny Pelaez',
    'Manuel Pelaez': 'Manny Pelaez'
})

df['FilerName'] = df['FilerName'].replace({
    'Manny Pelaez Pelaez': 'Manny Pelaez',
    'Manuel Pelaez': 'Manny Pelaez'
})

df['Cand/Committee:'].unique()
df['FilerName'].unique()
```

```python
df.to_csv('campaign_finance20240926.csv', index=False)
```
